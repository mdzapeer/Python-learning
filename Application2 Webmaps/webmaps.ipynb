```python
latlist=list(voldata["LAT"])
lonlist=list(voldata["LON"])
```

```python
map1=folium.Map(location=[38.58,-99.09], zoom_start=6, tiles="Mapbox Bright")
fg1=folium.FeatureGroup(name="My Map")
for lt,ln,name,loc in zip(latlist,lonlist,voldata.NAME, voldata.LOCATION):
    fg1.add_child(folium.CircleMarker(location = [lt, ln], popup=folium.Popup(name+" "+loc,parse_html=True), icon=folium.Icon(color='purple')))
    
map1.add_child(fg1)
```

```python
map1.save("map1a.html")
```

```python
map1=folium.Map(location=[38.58,-99.09], zoom_start=6, tiles="Mapbox Bright")
fg1=folium.FeatureGroup(name="My Map")
def elevcolor (elev):
    if elev < 1000:
        return "green"
    elif 1000<= elev <3000:
        return "orange"
    else:
        return "red"

for lt,ln,name,loc,elv in zip(voldata.LAT,voldata.LON,voldata.NAME, voldata.LOCATION, voldata.ELEV):
    fg1.add_child(folium.CircleMarker(location = [lt, ln],color=elevcolor(elv),popup=folium.Popup(name+" "+loc,parse_html=True)))

map1.add_child(fg1)
```
